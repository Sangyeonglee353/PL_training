#### 2. 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

- 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
- 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
- 반환하는 리스트이 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - div_path('/usr/local/bin/python')
        - ['/usr/local/bin', 'python']
    - div_path('/home/chulsoo/text.txt')
        - [/home/chulsoo', 'text.txt']
- [참고] 리스트(I) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 I.append(10)이다.

````python
def div_path(s): # 한 개의 문자열 인자를 받는 div_path() 함수 선언
    pathInfo = list() # 리스트 변수 선언
    index = s.rfind('/') # 문자열의 오른쪽부터 '/'가 처음나온 위치를 찾아 반환
    pathInfo.append(s[0:index]) # index를 경계로 앞은 디렉토리 경로명으로 리스트에 추가
    pathInfo.append(s[index+1:]) # index를 경계로 뒤는 파일명으로 리스트에 추가
    return pathInfo # 리스트값 반환
````

#### 3. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.

- 집합 자료형 set은 사용하지 않는다.
- 인자로 전달하는 리스트 2개에는 정수값만 들어간다고 가정하자.
- 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
- 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다(집합의 조건).
- 반환하는 리스트는 정렬이 되어 있어야 한다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - list_union([1, 2, 3], [1, 2, 4])   
        - [1, 2, 3, 4]
    - list_union([-10, -5, -1], [0, 9, 100])
        - [-10, -5, -1, 0, 9, 100]
    - list_union([0, 1, 2], [0, 1, 2])
        - [0, 1, 2]
- [참고] 리스트(I) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 I.append(10)이다.
- [참고] 임의의 정수값 (x)이 리스트(I) 내에 존재하는지 판단하는 방법은 x in I 이다.

````python
def list_union(lista, listb): # 두 개의 리스트를 인자로 받는 list_union() 함수 선언
    unionList = [] # 합집합을 저장한 리스트 선언
    for index in range(len(lista)): # 첫번째 리스트 추가
        if lista[index] not in unionList: # 리스트 원소와의 중복 여부 검사
            unionList.append(lista[index]) # 중복되지 않는다면 리스트에 추가
    for index in range(len(listb)): # 두번째 리스트 추가
        if listb[index] not in unionList: # 리스트 원소와의 중복 여부 검사
            unionList.append(listb[index]) # 중복되지 않는다면 리스트에 추가
    unionList.sort() # 합집합이 저장된 리스트를 오름차순 정렬
    return unionList # 최종 결과값 반환

# 고정된 값을 입력할 경우
print("# 리스트에 대한 합집합 출력")
print("list_union([1,2,3], [1,2,4]) --->", list_union([1,2,3], [1,2,4]))
print("list_union([-10, -5, -1], [0, 9, 100] --->", list_union([-10, -5, -1], [0, 9, 100]))
print("list_union([0,1,2], [0,1,2]) --->", list_union([0,1,2], [0,1,2]))
````

#### 4. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '교집합'을 반환하는 함수 list_intersection(lista, listb)와 '차집합'을 반환하는 함수 list_difference(lista, listb)를 작성하시오.

- 모든 가정과 조건은 3번 문제와 동일하다.

````python
def list_intersection(lista, listb): # 두 개의 리스트를 인자로 받는 list_intersection() 함수 선언
    templist = [] # 바로 비교시, 리스트의 크기에 따라 모든 원소를 비교 못하는 경우가 발생하기 때문에
                  # 비교를 위한 임시 리스트 선언
    intersectionList = [] # 교집합을 저장힐 리스트 선언
    for index in range(len(lista)): # 첫번째 리스트를 임시 리스트에 저장
        if lista[index] not in templist: # 리스트 원소와의 중복 여부 검사
            templist.append(lista[index]) # 중복되지 않는다면 리스트에 추가
    for index in range(len(listb)):  # 두번째 리스트의 원소가
        if listb[index] in templist: # 임시 리스트의 원소에 포함되어 있는지 검사
            intersectionList.append(listb[index]) # 원소가 포함되어 있다면 교집합 리스트에 추가
    intersectionList.sort() # 교집합이 저장된 리스트를 오름차순 정렬
    return intersectionList # 최종 결과값 반환

# 고정된 값을 입력할 경우
print("# 리스트에 대한 교집합 출력")
print("list_intersection([1,2,3], [1,2,4]) --->", list_intersection([1,2,3], [1,2,4]))
print("list_intersection([-10, -5, -1], [0, 9, 100]) -->", list_intersection([-10, -5, -1], [0, 9, 100]))
print("list_intersection([0,1,2], [0,1,2]) --->", list_intersection([0,1,2], [0,1,2]))
````

````python
def list_difference(lista, listb): # 두 개의 리스트를 인자로 받는 list_difference(lista, listb)
    differenceList = [] # 차집합을 저장할 리스트 선언
    for index in range(len(lista)): # 첫번째 리스트 추가
        if lista[index] not in differenceList: # 리스트 원소와의 중복 여부 검사
            differenceList.append(lista[index]) # 중복되지 않는다면 리스트에 추가
    for index in range(len(listb)):            # 두번째 리스트의 원소가
        if listb[index] in differenceList:     # 저장된 리스트에 포함되어 있다면
            differenceList.remove(listb[index]) # 해당 원소를 리스트에서 제거
    differenceList.sort() # 차집합이 저장된 리스트를 오름차순 정렬
    return differenceList # 최종 결과값 반환
    
# 고정된 값을 입력할 경우
print("# 리스트에 대한 차집합 출력")
print("list_difference([1,2,3], [1,2,4]) --->", list_difference([1,2,3], [1,2,4]))
print("list_difference([-10, -5, -1], [0, 9, 100]) -->", list_difference([-10, -5, -1], [0, 9, 100]))
print("list_difference([0,1,2], [0,1,2]) --->", list_difference([0,1,2], [0,1,2]))
````

#### 5. 두 개의 양의 정수를 인자로 받아서 해당 범위 안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a,b)를 작성하시오.

- 이미 구현된 모듈이나 라이브러리는 사용하지 않는다.
- 함수의 반환(return)은 없으며 함수 내의 끝에서 .format()을 사용하여 다음과 같이 출력한다.
    - print_primenumber(2, 10)
        - 2 이상 10 이하 정수 중 소수 리스트: [2, 3, 5, 7]

````python
def print_primenumber(a, b): # 두 개의 양의 정수를 인자로 받는 print_primenumber() 함수 선언
    if(a <= 0 or b <= 0): # 양의 정수가 아닌 0과 음수를 입력한 경우
        return print("잘못된 입력입니다. 양의 정수를 입력해주세요.") # 오류문 반환
    primeList = []                    # 소수를 저장할 리스트 선언
    check = 0                         # 소수 판별 기준: 소수인 경우 값은 0, 소수가 아닌 경우 값은 1
    for num1 in range(a, b+1):        # a부터 입력받은 값 b까지 반복
        for num2 in range(2, num1):   # 2부터 현재값 num1 전까지 반복
            if num1 % num2 == 0:      # 소수 판별: 본인이 아닌 다른 수와 나누어지면 소수가 아님
                check = 1             # 소수 판별 기준 1로 변경
        if check == 0 and num1 != 1: # 소수 판별 기준값이 0이고 숫자가 1이 아닌 경우 소수이므로 출력
            primeList.append(num1)
        check = 0                     # 다음값 판별을 위해 소수 판별 기준값을 0으로 초기화
    print("{} 이상 {} 이하 정수 중 소수 리스트:".format(a, b), primeList)
````

````python
import requests # html 소스를 가져오기 위해 삽입
import html # html escape 문자 변환을 위해 삽입

# 1. Google 소스 가져오기
req_google = requests.get('https://www.google.com/')
source_google = req_google.text

# 2. Naver소스 가져오기
req_naver = requests.get('https://www.naver.net/')
source_naver = req_naver.text

# 3. Kakao소스 가져오기
req_kakao = requests.get('https://www.kakaocorp.com/page/')
source_kakao = req_kakao.text
````

````python
# 추후 숙제 답안 제출 때엔 아래 코드는 주석처리할 것
# print(source_google)
````

````python
# 순수 문자열 텍스트만 걸러내는 stringFilter() 함수 선언
def stringFilter(source):
    wordList = []

    # 1. '<script ...> ~ </script>' 태그 내용 제거
    while((source.find('<script')+1) and (source.find('</script>')+1)): # 문자열.find()시 값이 없으면 -1이므로 True다.
                                                                         # 따라서 이 때 while()문 중단을 위해 +1을 하여
                                                                         # 값을 0으로 만들어 False를 통해 중단시킨다.
        firstIndex = source.find('<script')
        lastIndex = source.find('</script>') + len('</script>') # '</script>' 까지 제거해야 하므로
        source = source.replace(source[firstIndex:lastIndex], '')

    # 2. '<style ...> ~ </style>' 태그 내용 제거
    while((source.find('<style')+1) and (source.find('</style>')+1)): # 문자열.find()시 값이 없으면 -1이므로 True다.
                                                                       # 따라서 이 때 while()문 중단을 위해 +1을 하여
                                                                       # 값을 0으로 만들어 False를 통해 중단시킨다.
        firstIndex = source.find('<style') 
        lastIndex = source.find('</style>') + len('</style>') # '</style>' 까지 제거해야 하므로
        source = source.replace(source[firstIndex:lastIndex], '')

    # 3. '<'로 시작해서 '>' 끝나는 태그 제거
    while((source.find('<')+1) and (source.find('>')+1)): # 문자열.find()시 값이 없으면 -1이므로 True다.
                                                           # 따라서 이 때 while()문 중단을 위해 +1을 하여
                                                           # 값을 0으로 만들어 False를 통해 중단시킨다.
        firstIndex = source.find('<')
        lastIndex = source.find('>') + 1 # '>' 까지 제거해야 하므로
        source = source.replace(source[firstIndex:lastIndex], '')
    
    # 4. 제거가 끝난 source에서 html의 escape 문자를 문자로 변환
    source = html.unescape(source)
    
    # 5. 변환이 source를 리스트로 저장
    wordList = source.split()
    
    return wordList

# 필터링 결과 출력 함수 선언
def printResultFilter(source):
    print("1. 필터링된 source 값 리스트\n\n", stringFilter(source))
    print()
    print("2. 단어의 개수: {}개".format(len(stringFilter(source))))   
````

````python
# 구글 사이트 필터링 결과
printResultFilter(source_google)
````

````python
# 네이버 사이트 필터링 결과
printResultFilter(source_naver)
````

````python
# 카카오 사이트 필터링 결과
printResultFilter(source_kakao)
````
